```python
df1 = df.drop(['Transaction_ID','Date','Customer_ID','Age','Gender','Loyalty_Score','Signup_Date','Product_ID','City','Channel','Payment_Method',
            'Device','Promotion_Flag','Promo_Type','Margin_Percent','Return_Flag','Supplier_ID','Lead_Time_Days','Delivery_Delay_Days',
            'Stock_On_Hand','Marketing_Channel','Coupon_Code_Flag','Recency_Days','CLV_proxy','Frequency_12m','Online_Device','Season','Net_Revenue','Profit_Margin'], axis='columns')
df1
```

```python
dummies = pd.get_dummies(df1.Category)
dummies = dummies.astype(int)
print(dummies)
```

```python
df2 = pd.concat([df1,dummies.drop('Sports', axis='columns')] , axis='columns')
df2
```

```python
df3 = df2.drop(['Category'], axis='columns')
df3 = df2.drop(['Profit','GMV','Discount_Amount'])
X = df3
X.head(4)
```

```python
y = df['Revenue']
y.head(4)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
# K fold cross validation

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2 ,random_state=0)

cross_val_score(LinearRegression(), X,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor


def find_best_model_using_gridsearchcv(X,y):
    algo = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
            
            }
        },
        'Lasso' : {
            'model' : Lasso(),
            'params' : {
                 'alpha' : [1,2],
                'selection' : ['random','cyclic']
            }
        },
        'DecisionTreeRegressor' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['mse','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algo.items():
        gs = GridSearchCV(config['model'], config['params'], cv = cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model' : algo_name,
            'best_score' : gs.best_score_,
            'best_params' : gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_revenue(Category,Income,Price,Discount_Percent,Units_Sold,Cost,Profit, GMV, Discount_Amount):
    cat_index = np.where(X.columns==Category)[0][0]  # giving index to the hard coded category value

    x = np.zeros(len(X.columns))
    x[0] = Income
    x[1] = Price
    x[2] = Discount_Percent
    x[3] = Units_Sold
    x[4] = Cost
    x[5] = Profit
    x[6] = GMV
    x[7] = Discount_Amount
    if cat_index >= 0:
        x[cat_index] = 1
    return lr_clf.predict([x])[0]
predict_revenue('Beauty',78145,61.69,10,35,499.69,127.32,1110.42,222.08)
```
